```python
import graphviz
import numpy as np
import pandas as pd
from sklearn import tree
from sklearn import metrics
import matplotlib.pyplot as plt
from sklearn.tree import export_graphviz
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
```

```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("apeksha23/top-100-anime-dataset-ratings")

print("Path to dataset files:", path)
```

```python
df=pd.read_csv("C:/Users/yosef/.cache/kagglehub/datasets/apeksha23/top-100-anime-dataset-ratings/versions/3/anime_dataset.csv")

df.head()
```

Clean the data

```python
df = df.dropna()
df = df.drop(["Name", "Serial No.", "Links"], axis=1)

df.head()
```

Dummy values of genres and themes

```python
df = pd.concat([df, df['Genre and Theme'].str.get_dummies(sep=',').add_prefix('Genre: ')], axis=1)
df = df.drop('Genre and Theme', axis=1)
df.head()
```

Train/Test-split

```python
X = df.drop('Rating', axis=1)
y = df['Rating']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

Training of model and evaluation

```python
DT = DecisionTreeRegressor(random_state=42)
DT.fit(X_train, y_train)
y_pred = DT.predict(X_test)
score = mean_squared_error(y_test, y_pred)
print("Mean Squared Error: " + str(score))
```

```python
export_graphviz(DT, out_file='anime.dot', 
                class_names=True, 
                #class_names=['Negative', 'Positive'], 
                feature_names=df.columns.drop('Rating'),
                impurity=False,
                filled=True)
  
tree.plot_tree(DT)
```
